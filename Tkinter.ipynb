```python
import tkinter
from tkinter import *
from tkinter import messagebox

window = tkinter.Tk()
window.title("Result Calculator GUI")
window.geometry ("401x500")
window.resizable(0,0)

topic1 = Label(window, text = "Result Calculator GUI", font = ("Arial Bold",20),fg="#2dd902")
topic2 = Label(window, text = "Subject", font = ("Arial Bold",14),)
topic3 = Label(window, text = "Mark Obtained", font = ("Arial Bold",14))

mark1 = Label(window, text = " ", font = ("Arial Bold",12))
mark2 = Label(window, text = " ", font = ("Arial Bold",12))
mark3 = Label(window, text = " ", font = ("Arial Bold",12))
mark4 = Label(window, text = " ", font = ("Arial Bold",12))
perc =  Label(window, text = " ", font = ("Arial Bold",12))

grade =  Label(window, text = " ", font = ("Arial Bold",12))

phy = Label(window, text = "Physics", font = ("Arial Bold",12),fg="Red").place(x=0, y=82)
chem = Label(window, text = "Chemistry", font = ("Arial Bold",12),fg="Red").place(x=0, y=107)
math = Label(window, text = "Maths", font = ("Arial Bold",12),fg="Red").place(x=0, y=132)
eng = Label(window, text = "English", font = ("Arial Bold",12),fg="Red").place(x=0, y=156) 

phy_data = Entry(window, width="20")
chem_data = Entry(window, width="20")
math_data = Entry(window, width="20")
eng_data = Entry(window, width="20")

def clicked():
                                            # Getting mark for user and printing below submit button 
    try:
        phydata = float(phy_data.get())
        result1 = f"Physics mark:{phydata}"
        mark1.configure(text = result1, fg="Black")

        chemdata = float(chem_data.get())
        result2 = f"Chemistry mark: {chemdata}"
        mark2.configure(text = result2, fg="Black")    

        mathdata = float(math_data.get())
        result3 = f"Maths mark: {mathdata}"
        mark3.configure(text = result3, fg="Black")

        engdata = float(eng_data.get())
        result4 = f"English mark:{engdata}"
        mark4.configure(text = result4, fg="Black")


                                             # Failed when mark is less than 40
        if phydata < 40:
            Remark = f"Physics mark: {phydata} (F)"
            mark1.configure(text = Remark,fg="Red")
        if chemdata < 40:
            Remark = f"Chemistry mark: {chemdata} (F)"
            mark2.configure(text = Remark,fg="Red")
        if mathdata < 40:
            Remark = f"Maths mark: {mathdata} (F)"
            mark3.configure(text = Remark,fg="Red")
        if engdata < 40:
            Remark = f"English mark: {engdata} (F)"
            mark4.configure(text = Remark,fg="Red")
            
                                            # Invalid mark when mark is greater than 100 or smaller than 0
        if phydata > 100 or phydata < 0:
            Remark = f"Physics mark: Invalid"
            mark1.configure(text = Remark,fg="Red")
        if chemdata > 100 or chemdata < 0:
            Remark = f"Chemistry mark: Invalid"
            mark2.configure(text = Remark,fg="Red")
        if mathdata > 100 or mathdata < 0:
            Remark = f"Maths mark: Invalid"
            mark3.configure(text = Remark,fg="Red")
        if engdata > 100 or engdata < 0:
            Remark = f"English mark: Invalid"
            mark4.configure(text = Remark,fg="Red")

                                            # Calculating Percentage by converting user-string into float
            
        perc_int = ( float(phy_data.get()) + float(chem_data.get()) + float(math_data.get()) + float(eng_data.get()) ) / 4
        Final = f"Percentage:{perc_int}% "
        perc.configure(text = Final, fg="Green")


                                            # Grading According to percentage

        if perc_int <= 100 and perc_int >= 90:
            Grade = "Grade: A+"
            grade.configure(text = Grade,fg="Green")
        elif perc_int <= 89 and perc_int >= 80:
            Grade = "Grade: A"
            grade.configure(text = Grade,fg="Green")
        elif perc_int <= 79 and perc_int >= 70:
            Grade = "Grade: B+"
            grade.configure(text = Grade,fg="Green")
        elif perc_int <= 69 and perc_int >= 60:
            Grade = "Grade: B"
            grade.configure(text = Grade,fg="Green")
        elif perc_int <= 59 and perc_int >= 50:
            Grade = "Grade: C+"
            grade.configure(text = Grade,fg="Green")
        elif perc_int <= 49 and perc_int >= 40:
            Grade = "Grade: C"
            grade.configure(text = Grade,fg="Green")
        elif perc_int <40:
            Grade = "Grade: F- Failed"
            grade.configure(text = Grade,fg="red")
            perc.configure(fg="Red")
            
                                            # Invalid Grading/Percentage when mark is above 100 or negative
                
        if perc_int > 100 or phydata > 100 or chemdata > 100 or mathdata > 100 or engdata > 100:
            Grade = "Grade: Invalid"
            grade.configure(text = Grade,fg="Blue")

            Percentage = "Percentage: Invalid" 
            perc.configure(text = Percentage,fg="Blue")

        if perc_int < 0 or phydata < 0 or chemdata < 0 or mathdata < 0 or engdata < 0:
            Grade = "Grade: Invalid"
            grade.configure(text = Grade,fg="Blue")
            
            Percentage = "Percentage: Invalid" 
            perc.configure(text = Percentage,fg="Blue")
            
                                            # Error message when no input
            
    except:
        error = messagebox.showerror("Error", "Invalid Input. Check Again !!")
        
        
                                            # Checking whether the checkbox is ticked or not
def chk():       
    if chk_state.get() == 1:
        button['state'] = NORMAL
        button.configure(text="Submit Marks")

    elif chk_state.get() == 0:
        button['state'] = DISABLED
        button.configure(text="Tick to submit")

chk_state = IntVar()
Check = Checkbutton(window, text = "Above marks are correctly entered.", var = chk_state, onvalue=1, offvalue=0, command=chk)

button = Button(window, text= "Tick to submit",font = ("Arial Bold",10),fg="Red", state = DISABLED,command = clicked)


topic1.grid(column =1, row = 0)
topic2.grid(column =0, row =1, pady = 10)
topic3.grid(column =1, row =1)

phy_data.grid(column =1, row =2)
chem_data.grid(column =1, row =3,pady = 5)
math_data.grid(column =1, row =4)
eng_data.grid(column =1, row =5,pady = 5)

Check.grid(column = 1, row = 6, pady = 10)

button.grid(column =1, row =7)

mark1.grid(column =1, row =8,pady = 5)
mark2.grid(column =1, row =9)
mark3.grid(column =1, row =10,pady = 5)
mark4.grid(column =1, row =11)
perc.grid(column =1, row =12,pady = 5)
grade.grid (column = 1, row = 13)


window.mainloop()
```
